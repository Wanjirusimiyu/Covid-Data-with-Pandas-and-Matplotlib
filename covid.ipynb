```python
# Only drop rows where 'date', 'new_cases', or 'location' is missing
df_filtered = df.dropna(subset=['date', 'new_cases', 'location'])

df_filtered['date'] = pd.to_datetime(df_filtered['date'])

# List of countries
countries = ['Kenya', 'Uganda', 'Tanzania']

# Get only those rows where 'location' is in that list
multi_countries = df_filtered[df_filtered['location'].isin(countries)]


plt.figure(figsize=(12, 6))

# Loop through each country and plot its data
for country in countries:
    country_data = multi_countries[multi_countries['location'] == country]
    plt.plot(country_data['date'], country_data['new_cases'], label=country)

plt.title('New COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


```

```python
df_cases_deaths = df[['new_cases', 'new_deaths']].dropna()
# Set seaborn style
sns.set(style="whitegrid")

plt.figure(figsize=(12, 8))

# Plot the scatter plot
sns.scatterplot(
    x='new_cases',
    y='new_deaths',
    data=df_cases_deaths,
    color='mediumseagreen',
    alpha=0.5,
    edgecolor='black'
)

# Titles and labels
plt.title('Relationship Between New COVID-19 Cases and Deaths', fontsize=16, fontweight='bold')
plt.xlabel('New Cases', fontsize=14)
plt.ylabel('New Deaths', fontsize=14)

# Optional: Limit the axes to zoom in
plt.xlim(0, 10000)
plt.ylim(0, 500)

# Add grid for easier reading
plt.grid(True, linestyle='--', alpha=0.6)

# Show the plot
plt.show()


```

# My Observations

**1. Average New Cases by Continent**

From the bar chart, we can see that Europe and Asia had the highest average number of new COVID-19 cases compared to other continents. This might be due to large population sizes and urban areas where viruses spread faster.

**2. New Cases Over Time in Kenya**

The line chart shows clear waves of COVID-19 infections in Kenya, with spikes during certain months. This could represent different waves of the pandemic or major public events that led to a rise in cases.

**3. Histogram of New Deaths**

Most of the data points in the histogram are clustered near zero, meaning that many days had few or no new deaths. However, there are some days with very high death numbers, possibly during severe waves.

**4. New Cases vs New Deaths (Scatter Plot)**

The scatter plot shows a positive relationship: when the number of new cases increases, the number of new deaths also tends to increase. But there are also many days with high cases but low deaths, which could be due to better treatment or vaccination.
